```python
from Supportingcode.Data_reader import Data
from Supportingcode.Algorithms import rescheduling_problem
import os
import igraph
import pandas as pd
```

## 1 Hour instances

```python
local = 'Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/'
logfilepath = local+'LogFiles/'
local2 = 'Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/CSVFiles/'
arquivos = os.listdir(local2)
results1 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "Termination condition", "number pf binary variables"])
results2 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "Termination condition", "number pf binary variables"])
# results3 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "# of routes","Termination condition", "number pf binary variables"])
results3 = {}
for a in arquivos:
    print('\n',a)
    info = Data(a, local, "edb")
    tracks, stations, trains, current_time, maxtime = info.read_from_CSV()
    rescheduling = rescheduling_problem(tracks, stations, trains, current_time, maxtime, a)
    Fixedroute_reschedule1 = rescheduling.Model1(False, logfilepath)
    results1.loc[len(results1)] = [a, Fixedroute_reschedule1[0], Fixedroute_reschedule1[1], Fixedroute_reschedule1[2], Fixedroute_reschedule1[3], Fixedroute_reschedule1[4]]
    Fixedroute_reschedule2 = rescheduling.Model2(False, logfilepath)
    results2.loc[len(results2)] = [a, Fixedroute_reschedule2[0], Fixedroute_reschedule2[1], Fixedroute_reschedule2[2], Fixedroute_reschedule2[3], Fixedroute_reschedule2[4]]
    for r in [2,10]:
        if not(r in results3.keys()):
            results3[r] = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "# of routes","Termination condition", "number pf binary variables"])        
        # if r>1 and results3[r-1][results3[r-1]["Data"]==a]["# of routes"][0] == results3[r-2][results3[r-2]["Data"]==a]["# of routes"][0]:
        #     pass
        # else:
        #     Alternativeroute_reschedule = rescheduling.Model3(False, r+1, logfilepath)
        #     results3[r].loc[len(results3[r])] = [f'{a}', Alternativeroute_reschedule[0], Alternativeroute_reschedule[1], Alternativeroute_reschedule[2], Alternativeroute_reschedule[3], Alternativeroute_reschedule[4], Alternativeroute_reschedule[5]]
        Alternativeroute_reschedule = rescheduling.Model3(False, r, logfilepath)
        results3[r].loc[len(results3[r])] = [f'{a}', Alternativeroute_reschedule[0], Alternativeroute_reschedule[1], Alternativeroute_reschedule[2], Alternativeroute_reschedule[3], Alternativeroute_reschedule[4], Alternativeroute_reschedule[5]]
pd.DataFrame().to_excel("Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/results.xlsx")
with pd.ExcelWriter("Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/results.xlsx", engine='openpyxl', mode='a') as writer:
    results1.to_excel(writer, sheet_name="Model1", index=False)
    results2.to_excel(writer, sheet_name="Model2", index=False)
    for r in results3:
        results3[r].to_excel(writer, sheet_name=f"Model3{r}", index=False)
```

## Complete schedule instance

```python
local = 'Data/complete-madrid-instance/'
logfilepath = local+'LogFiles/'
results1 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "Termination condition", "number pf binary variables"])
results2 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "Termination condition", "number pf binary variables"])
results3 = {}
info = Data("madrid_current_schedule.edb", local, "edb")
tracks, stations, trains, current_time, maxtime = info.read_from_CSV()
a = "Complete schedule"
rescheduling = rescheduling_problem(tracks, stations, trains, current_time, maxtime, a)
Fixedroute_reschedule1 = rescheduling.Model1(False, logfilepath)
results1.loc[len(results1)] = [a, Fixedroute_reschedule1[0], Fixedroute_reschedule1[1], Fixedroute_reschedule1[2], Fixedroute_reschedule1[3], Fixedroute_reschedule1[4]]
Fixedroute_reschedule2 = rescheduling.Model2(False, logfilepath)
results2.loc[len(results2)] = [a, Fixedroute_reschedule2[0], Fixedroute_reschedule2[1], Fixedroute_reschedule2[2], Fixedroute_reschedule2[3], Fixedroute_reschedule2[4]]
for r in range(3):
    results3[r+1] = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "# of routes","Termination condition", "number pf binary variables"])
    Alternativeroute_reschedule = rescheduling.Model3(False,r+1, logfilepath)
    results3.loc[len(results3)] = [a, Alternativeroute_reschedule[0], Alternativeroute_reschedule[1], Alternativeroute_reschedule[2], Alternativeroute_reschedule[3], Alternativeroute_reschedule[4], Alternativeroute_reschedule[5]]
pd.DataFrame().to_excel("Data/complete-madrid-instance/results.xlsx")
with pd.ExcelWriter("Data/complete-madrid-instance/results.xlsx", engine='openpyxl', mode='a') as writer:
    results1.to_excel(writer, sheet_name="Model1", index=False)
    results2.to_excel(writer, sheet_name="Model2", index=False)
    for r in results3:
        results3[r].to_excel(writer, sheet_name=f"Model3{r}", index=False)
```
