```python
from Supportingcode.Data_reader import Data
from Supportingcode.Algorithms import rescheduling_problem
import os
import igraph
import pandas as pd
```

```python
local = 'Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/'
local2 = 'Data/1-hour-interval-madrid/1-hour-interval-madrid/testsuite/CSVFiles/'
arquivos = os.listdir(local2)
results1 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time"])
results2 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time"])
results3 = pd.DataFrame(columns=["Data", "# of trains", "Objective value", "solution time", "# of routes"])
for a in arquivos:
    print('\n',a)
    info = Data(a, local, "edb")
    tracks, stations, trains, current_time, maxtime = info.read_from_CSV()
    rescheduling = rescheduling_problem(tracks, stations, trains, current_time, maxtime, a)
    Fixedroute_reschedule1 = rescheduling.Model1()
    results1.loc[len(results1)] = [a, Fixedroute_reschedule1[0], Fixedroute_reschedule1[1], Fixedroute_reschedule1[2]]
    Fixedroute_reschedule2 = rescheduling.Model2()
    results2.loc[len(results2)] = [a, Fixedroute_reschedule2[0], Fixedroute_reschedule2[1], Fixedroute_reschedule2[2]]
    Alternativeroute_reschedule = rescheduling.Model3()
    results3.loc[len(results3)] = [a, Alternativeroute_reschedule[0], Alternativeroute_reschedule[1], Alternativeroute_reschedule[2], Alternativeroute_reschedule[3]]
pd.DataFrame().to_excel("Data/results.xlsx")
with pd.ExcelWriter("Data/results.xlsx", engine='openpyxl', mode='a') as writer:
    results1.to_excel(writer, sheet_name="Model1", index=False)
    results2.to_excel(writer, sheet_name="Model2", index=False)
    results3.to_excel(writer, sheet_name="Model3", index=False)
```
